```python
import csv
import nltk
from nltk.tokenize import sent_tokenize

# Download NLTK punkt tokenizer (if not already downloaded)
nltk.download('punkt')

def split_sentences(text):
    sentences = sent_tokenize(text)
    return sentences

# Read Harry Potter text from file
input_file_path = "harrypotter.txt"
output_csv_file_path = "harrypotter.csv"

try:
    with open(input_file_path, 'r', encoding='utf-8') as file:
        harry_potter_text = file.read()

    # Split the text into sentences
    sentences = split_sentences(harry_potter_text)

    # Remove newlines from each sentence
    sentences = [sentence.replace('\n', ' ') for sentence in sentences]

    # Write sentences to a CSV file
    with open(output_csv_file_path, 'w', encoding='utf-8', newline='') as csv_file:
        csv_writer = csv.writer(csv_file)
        csv_writer.writerow(['Sentence'])

        for sentence in sentences:
            csv_writer.writerow([sentence])

    print(f"Sentences written to '{output_csv_file_path}'")

except FileNotFoundError:
    print(f"Error: File '{input_file_path}' not found.")
except Exception as e:
    print(f"Error: {e}")
```
